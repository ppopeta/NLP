```python
import tensorflow as tf
from tensorflow import keras

frases = [
    'Hola mundo',
    'Hola a todos',
    'Hola a todo el mundo',
    'Buenos dias, ¿como estas hoy?'
]

#Generar l diccionario de tokens 
print("-------------------------------")

tokenizer = keras.preprocessing.text.Tokenizer(num_words=10)
tokenizer.fit_on_texts(frases)
word_index = tokenizer.word_index
print(f"Word index: {word_index}")
print("-------------------------------")

#Generacion de secuencias de tokens
secuencias = tokenizer.texts_to_sequences(frases)
print(f"Secuencias: {secuencias}")
print("-------------------------------")


#Rellena las secunecias para que tengan la misma longitud
relleno = keras.preprocessing.sequence.pad_sequences(secuencias)
print(f"Secuencias rellenas:\n {relleno}")
print("-------------------------------")


```
